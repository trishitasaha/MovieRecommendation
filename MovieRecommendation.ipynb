```python
import pandas as pd 
```

```python
import numpy as np 
```

```python
#Read the dataset of movies 
a=pd.read_csv('tmdb_5000_movies.csv') 
```

```python
#Read the dataset of movie credits 
b=pd.read_csv('tmdb_5000_credits.csv') 
```

```python
#Display the name of columns of first dataset
a.columns 
```

```python
#Display the name of columns of second dataset 
b.columns 
```

```python
#Merge the two datasets w.r.t. title 
data=pd.merge(a, b, on='title') 
```

```python
#Display the name of columns of the merged dataset 
data.columns 
```

```python
#Print the first 100 recommended movie names based on credits 
data.groupby('title')['vote_average'].mean().sort_values(ascending=False).head(100) 
```
